Created a dictionary for storing the data

```python
Dict={"akash":10000,"rohit":20000,"ankit":25000,"akshay":30000,"nina":15000}
```

Created differnt functions to run the program

```python
#Account creation function
#account creation
def acct_crt():
    x=input("Enter the Name :").lower()
    if x in Dict.keys():
        print("Account already existed!")
    else: 
        y=int(input("Enter the amount you want to deposite :"))
        if y>=10000:
            pass
        else:
            print("Minimum amount to open the bank balance is 10000 ","|","Please enter the appropriate amount")
            return

        
        Dict.update({x:y})
        print("Account created successfully")
        print("Want to add balance to your newly opened account if yes, then enter","yes","if no, then enter","no")
        a=input("Enter your choice :").lower()
        if a=="yes":
            acct_dpst()
        else:
            print("Thank you")
            return
        
    return
#account balance check
def bal_chk():  
    x=input("Enter the Name :").lower()
    if x in Dict.keys():
        a=Dict.get(x)
        print("Account Balance available is",a)
    else:
        print("Account do not exist!")
    return
#account withdrawal function
def acct_wdr():
    x=input("Enter the Name :").lower()
    if x in Dict.keys():
        print("Acount exist")    
        y=int(input("Enter the amount you want to withdraw :"))
        if Dict.get(x)<=100:
            print("Not enough balance present!","|","Account Balance is",Dict.get(x))
            return
        else:
            if Dict.get(x)>=y:
                a=Dict.get(x)-y
        
                Dict.update({x:a})
    
                print("Updated balance is",a)
            else:
                print("Balance not sufficient!","|","Available balance is",Dict.get(x))
                return
            return
        return

    else:
        print("Account do not exist!")
        return
    return
#balance deposite
def acct_dpst():
    x=input("Enter the Name :").lower()
    if x in Dict.keys():
        print("Acount exist")    
        y=int(input("Enter the amount you want to deposite :"))
        if y<=0:
            print("Enter a valid amount!")
            return
        else:
            pass
        if y>=100000:
            print("You cannot deposite amount greater than or equal to 100000","|","Please try again with lesser amount")
            return
        else:
            
            a=Dict.get(x)+y
        
            Dict.update({x:a})
    
            print("Updated balance is",Dict.get(x))
        return
    else:
        print("Account do not exist!")

print("Welcome to the XYZ bank")
print("""Please select the option you want to continue with
        1.Account Creation
        2.Balance Check
        3.Withdraw
        4.Deposite """)
a=int(input("Please,enter your Choice :"))
if a==1:
    acct_crt()
elif a==2:
    bal_chk()
elif a==3:
    acct_wdr()
elif a==4:
    acct_dpst()
else:

    print("Enter correct input")
```
